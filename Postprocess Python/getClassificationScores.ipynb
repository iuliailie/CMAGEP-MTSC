```python
realClass=[]
predictedClass=[]
accuracies=[]
Modelcombo=[]
f1_scores=[]
precisions=[]
totalIndex=0
MeanCOD=[]
for c1_index in range(0,dfResultsC1.shape[0]):

    for c2_index in range(0,dfResultsC2.shape[0]):
        cmagep_c1=dfResultsC1['optimized structure'].iloc[c1_index]
        cmagep_c2=dfResultsC2['optimized structure'].iloc[c2_index]


        realClassinter=[]
        coefficient_of_dermination1=[]
        coefficient_of_dermination2=[]
        predictedClassInter=[]       
        gb=Xtest.groupby(Xtest.columns[-1])
        gbY=Ytest.groupby(Ytest.columns[-1])
        for i in range(0,len(gb)):
            da=[]
            df=gb.get_group(i)
            dfY=gbY.get_group(i)
            
               # z=df['s24']

            realClassinter.append(df[df.columns[-2]].iloc[0])
            

            try:
                cmagep_c1_=cmagep_c1.replace('^','**')
                cmagep_c1_=cmagep_c1_.replace('exp','np.exp')
                cmagep_c1_=cmagep_c1_.replace('log','np.log')
                indCols=len(newCols)-3
                while (indCols>-1):
                    cmagep_c1_=cmagep_c1_.replace(newCols[indCols],'df[df.columns['+str(indCols)+']]')
                    indCols=indCols-1
                            
                c1_Pred=eval(cmagep_c1_)

                coefficient_of_dermination1.append(r2_score(dfY[dfY.columns[0]],c1_Pred ))
            except:
                coefficient_of_dermination1.append(-100)
            try:
                cmagep_c2_=cmagep_c2.replace('^','**')
                cmagep_c2_=cmagep_c2_.replace('exp','np.exp')
                cmagep_c2_=cmagep_c2_.replace('log','np.log')
                indCols=len(newCols)-3
                while (indCols>-1):
                    cmagep_c2_=cmagep_c2_.replace(newCols[indCols],'df[df.columns['+str(indCols)+']]')
                    indCols=indCols-1
                    
                c2_Pred=eval(cmagep_c2_)
                coefficient_of_dermination2.append(r2_score(dfY[dfY.columns[0]],c2_Pred))
            except  :          
                coefficient_of_dermination2.append(-100)

            if coefficient_of_dermination1[i]>coefficient_of_dermination2[i]:
                predictedClassInter.append(PossibleClasses[0])
            else:
                predictedClassInter.append(PossibleClasses[1])

        rc=np.asarray(realClassinter)
       # rc[rc>1]=0
        realClass.append(rc)
        predictedClass.append(np.asarray(predictedClassInter))
        f1_scores.append(f1_score(realClass[totalIndex],predictedClass[totalIndex],average='binary'))

        accuracies.append(accuracy_score(realClass[totalIndex],predictedClass[totalIndex]))
        precisions.append(precision_score(realClass[totalIndex],predictedClass[totalIndex]))
        Modelcombo.append([c1_index,c2_index])
        MeanCOD.append(np.mean(coefficient_of_dermination1))
        totalIndex=totalIndex+1

maxIndex=MeanCOD.index(max(MeanCOD))
```
